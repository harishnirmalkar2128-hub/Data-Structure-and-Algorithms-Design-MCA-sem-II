```python
def evaluate_postfix(exp):
    stack = []
    for char in exp:
        if char.isdigit():
            stack.append(int(char))
        else:
            val1 = stack.pop()
            val2 = stack.pop()
            if char == '+': stack.append(val2 + val1)
            elif char == '-': stack.append(val2 - val1)
            elif char == '*': stack.append(val2 * val1)
            elif char == '/': stack.append(val2 / val1)
    return stack.pop()

print("Result of '231*+9-':", evaluate_postfix("231*+9-"))
```
