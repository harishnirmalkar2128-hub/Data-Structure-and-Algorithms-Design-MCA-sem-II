```python
# 2D Matrix (3x2)
matrix = [
    [10, 20],
    [30, 40],
    [50, 60]
]

# Row-Major conversion logic
row_major = []
for row in matrix:
    for element in row:
        row_major.append(element)

print("Original Matrix:", matrix)
print("Row-Major Order:", row_major) 
# Output: [10, 20, 30, 40, 50, 60]
```
