#Program to Search an Element in a List

```python
# Search element in a list

lst = [5, 15, 25, 35, 45, 55, 65, 75, 85, 95]

print("List:", lst)

key = int(input("Enter element to search: "))

if key in lst:
   print("Element found at index:", lst.index(key))
else:
   print("Element not found")
```
