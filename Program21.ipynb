```python
import heapq

def prims(graph, start):
    mst = []
    visited = {start}
    edges = [(w, start, to) for to, w in graph[start]]
    heapq.heapify(edges)
    
    while edges:
        w, frm, to = heapq.heappop(edges)
        if to not in visited:
            visited.add(to)
            mst.append((frm, to, w))
            for next_node, weight in graph[to]:
                if next_node not in visited:
                    heapq.heappush(edges, (weight, to, next_node))
    return mst

adj = {
    0: [(1, 2), (3, 6)],
    1: [(0, 2), (2, 3), (3, 8), (4, 5)],
    2: [(1, 3), (4, 7)],
    3: [(0, 6), (1, 8)],
    4: [(1, 5), (2, 7)]
}
print("Prim's MST (from, to, weight):", prims(adj, 0))
```
