```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

head = Node(1)
second = Node(2)
third = Node(3)
fourth = Node(4)
fifth = Node(5)
sixth = Node(6)
seventh = Node(7)
eighth = Node(8)
ninth = Node(9)
tenth = Node(10)


head.next = second
second.next = third 
third.next = fourth
fourth.next = fifth
fifth.next = sixth  
sixth.next = seventh
seventh.next = eighth
eighth.next = ninth
ninth.next = tenth


temp = head
while temp:
   print(temp.value)
   temp = temp.next
   
print("End of linked list")
```
