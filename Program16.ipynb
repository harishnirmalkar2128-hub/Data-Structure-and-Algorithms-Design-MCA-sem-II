```python
import time, random

# 1. Selection Sort Logic
def selection_sort(a):
    for i in range(len(a)):
        m = i
        for j in range(i+1, len(a)):
            if a[j] < a[m]: m = j
        a[i], a[m] = a[m], a[i]

# 2. Insertion Sort Logic
def insertion_sort(a):
    for i in range(1, len(a)):
        k = a[i]; j = i-1
        while j >= 0 and k < a[j]: a[j+1] = a[j]; j -= 1
        a[j+1] = k

# 3. Merge Sort Logic
def merge_sort(a):
    if len(a) > 1:
        m = len(a)//2; L = a[:m]; R = a[m:]
        merge_sort(L); merge_sort(R)
        i = j = k = 0
        while i<len(L) and j<len(R):
            if L[i]<R[j]: a[k]=L[i]; i+=1
            else: a[k]=R[j]; j+=1
            k+=1
        a[k:] = L[i:] + R[j:]

# --- Comparison Code ---
data = [random.randint(1, 1000) for _ in range(1000)]
print(f"Comparing algorithms with {len(data)} elements...\n")

for name, func in [("Selection Sort", selection_sort), 
                   ("Insertion Sort", insertion_sort), 
                   ("Merge Sort", merge_sort)]:
    test_data = data[:]
    start = time.time()
    func(test_data)
    print(f"{name}: {time.time() - start:.6f} seconds")
```
