```python
class Node:
    def __init__(self, key):
        self.left = self.right = None
        self.val = key

def insert(root, key):
    if root is None:
        return Node(key)
    if key < root.val:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)
    return root

def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

r = Node(50)
elements = [30, 20, 40, 70, 60, 80]
for x in elements:
    insert(r, x)

print("Binary Search Tree created.")
print("Inorder Traversal (Sorted Order):", end=" ")
inorder(r)
print()
```
