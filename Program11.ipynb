```python
# 2D Matrix (3x2)
matrix = [
    [10, 20],
    [30, 40],
    [50, 60]
]

rows = len(matrix)       # 3
cols = len(matrix[0])    # 2

# Column-Major conversion logic
col_major = []
for j in range(cols):
    for i in range(rows):
        col_major.append(matrix[i][j])

print("Original Matrix:", matrix)
print("Column-Major Order:", col_major)
# Output: [10, 30, 50, 20, 40, 60]
```
